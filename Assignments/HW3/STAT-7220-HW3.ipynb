7.  Fit a full factorial ANOVA model to the data. Assess the normality assumption using both a visual method and a testing method. Do we have evidence to support the assumption of normality? Why or why not?

* On the QQ plot we can see that the majority of the data point lie on the 45 degree line bisecting the graph with minor devations of approximately 10 percent at the top and bottom ends of the line. This supports the assumption of normality. Additionaly, from the Shapiro-Wilk test we calculate a p-value=0.429 which fails to reject the null hypothesis at an alpha of 0.05. It follows that the statistical test also supports the assumption of normality.


```python
## Import Statsmodels ##
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy import stats

## Fit Full Factorial Model ##

model = ols('comfort ~ sole_material*insole_thickness*breathability', data=shoe).fit()

## Parse out model residuals ##

residuals = model.resid

## Test of Normality ##

## QQ-Plot ##

sm.qqplot(residuals, line='s')
plt.title("QQ-Plot of Model Residuals")
plt.show()

## Shapiro-Wilk Test ##

shapiro_test = stats.shapiro(residuals)

print("Shapiro-Wilk Test Results:")
print(f"Statistic: {shapiro_test.statistic}")
print(f"p-value: {shapiro_test.pvalue}")
if shapiro_test.pvalue < 0.05:
    print("Reject the null hypothesis - Residuals are not normally distributed.")
else:
    print("Fail to reject the null hypothesis - Residuals are normally distributed.")
```


8.  Assess the constant variance assumption using both a visual method and a testing method. Do we have evidence to support the assumption of constant variance? Why or why not?

* In the fitted vs. residual plot we can see a random scattering of data points with no clear patterns of coning or clustering. The data points are generally centered around a mean of zero. The banding in the data can be explained by the categorical nature of the variables. Additionally, the Breuch-Pagan test fails to reject the null hypothesis with a p-value=0.083, this further supports the assumption of homoscedasticity.


```python
## Test of Homogeneity of Variance ##

## Scatterplot of Standardized Residuals vs. Fitted Values ##

## Get Standardized Residuals ##

standardized_residuals = model.get_influence().resid_studentized_internal

## Get Fitted Values ##

fitted_values = model.fittedvalues

## Scatterplot ##

## Create the plot ##

plt.figure(figsize=(8, 6))
sns.scatterplot(x=fitted_values, y=standardized_residuals)
plt.axhline(y=0, linestyle='dashed', color='red')
plt.axhline(y=3, color='blue')
plt.axhline(y=-3, color='blue')
plt.xlabel("Predicted y's")
plt.ylabel("Standardized Residuals")
plt.title("Standardized Residuals vs. Fitted Values")
plt.style.use('classic')
plt.show()

## B-P Test ##

## Run B-P Test ##

bp_test = het_breuschpagan(residuals, model.model.exog)

## Extract F-statistic ##

f_statistic = bp_test[2]

## Extract its P-Value ##

f_p_value = bp_test[3]

print(f"F-statistic: {f_statistic}")
print(f"F p-value: {f_p_value}")
if f_p_value < 0.05:
    print("Reject the null hypothesis - Variances are not equal.")
else:
    print("Fail to reject the null hypothesis - Variances are equal.")
```

```python
## First, import ols function ##

from statsmodels.formula.api import ols
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

## Fit the Model

fitbit_mod = ols('Satisfaction ~ GPS_Accuracy + Battery_Life + Display', data=fitness).fit()
summary = fitbit_mod.summary()
print(summary)

## Test of Normality ##
## Extract Model Residuals ##

residuals = fitbit_mod.resid

## Use statsmodels to plot qqplot ##

sm.qqplot(residuals, line ='45')
plt.title('Q-Q Plot of Model Residuals')
plt.show()

## Run Shapiro-Wilk Test for Normality ##

shapiro_test = stats.shapiro(residuals)
print(f'Shapiro-Wilk Test Statistic: {shapiro_test.statistic}')
print(f'p-value: {shapiro_test.pvalue}')
if shapiro_test.pvalue < 0.05:
    print("Reject the null hypothesis: Residuals are not normally distributed.")
else:
    print("Fail to reject the null hypothesis: Residuals are normally distributed.")
```

```python
## Group the Data by the Effect column and pivot from long to wide ##

effect_map = {1: 'c', 2: 'a', 3: 'b', 4: 'abc'}
fitness['Effect'] = fitness['Run_ID'].map(effect_map)

## Group the Data by the Effect column and pivot from long to wide ##
df = (fitness.groupby("Effect", as_index=False)["Satisfaction"].mean())


df = df.pivot_table(index=None, columns='Effect', values='Satisfaction').reset_index(drop=True)

## Extract factor columns for easier use ##
a   = df['a']
b   = df['b']
c   = df['c']
abc = df['abc']

## Compute factor estimates ##
df2 = pd.DataFrame({
    "A (GPS)": 0.5 * (a - b - c + abc),
    "B (Battery)": 0.5 * (-a + b - c + abc),
    "C (Display)": 0.5 * (-a - b + c + abc)
})

print("Average Satisfaction per Effect:\n", df, "\n")
print("Main Effect Estimates:\n", df2)
```

12. Calculate and interpret the partial $\eta^2$ values for each of the main effects. Which effects are most important? Why?
* Display and GPS Accuracy are the two most important main effects accoutning for approximately 95% and 93% of variance respectively. Battery life is also important but to a lesser degree than the other two, accounting for approximately 71% of variance.

```python

## Calculate Partial Eta Squared for Each Factor ##

## Load pingouin ##

import pingouin as pg

## Fit ANOVA with pingouin ##

mod2 = pg.anova(data=fitness, dv='Satisfaction', between=['GPS_Accuracy','Battery_Life','Display'], detailed=True)

print(mod2[["Source","np2"]][:3])
```

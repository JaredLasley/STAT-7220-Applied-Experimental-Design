```python
## Import Relevant Packages ##
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

## Read in burrito Data ##

burrito = pd.read_excel("Busters Burritos.xlsx")

## Data Integrity Check ##

print(burrito.info())

## Exploratory Data Analysis ##
## Start with Main Effects ##

## Protein ##

protien_stats = burrito.groupby("Protein")["Satisfaction"].agg(['mean', 'std'])

print(protien_stats)

## Salsa ##

Salsa_stats = burrito.groupby("Salsa")["Satisfaction"].agg(['mean', 'std'])

print(Salsa_stats)

## Potatoes ##

breath_stats = burrito.groupby("Potatoes")["Satisfaction"].agg(['mean', 'std'])

print(breath_stats)

#two-way interactions
## Protein x Salsa ##

sole_insole_stats = burrito.groupby(["Protein", "Salsa"])["Satisfaction"].agg(['mean', 'std'])

print(sole_insole_stats)

## Protein x Potatoes ##

sole_breath_stats = burrito.groupby(["Protein", "Potatoes"])["Satisfaction"].agg(['mean', 'std'])

print(sole_breath_stats)

## Salsa x Potatoes ##

insole_breath_stats = burrito.groupby(["Salsa", "Potatoes"])["Satisfaction"].agg(['mean', 'std'])

print(insole_breath_stats)

#three-way interactions
sole_insole_breath_stats = burrito.groupby(["Protein", "Salsa", "Potatoes"])["Satisfaction"].agg(['mean', 'std'])

print(sole_insole_breath_stats)

#plotting the data
## Install plotnine
#%pip install plotnine
## Import all plotnine functions ##:
from plotnine import *

# Prepare summarized data
burrito_means = sole_insole_breath_stats.reset_index()[["Protein", "Salsa", "Potatoes", "mean"]]

# Define dodge position for aligning points/lines with the boxplots
box_dodge = position_dodge(0.75)

# Build a single faceted plot
p = (
    ggplot(burrito, aes(x='Protein', y='Satisfaction')) +

    # 1. Boxplot layer (base visualization)
    geom_boxplot(aes(fill='Salsa'), alpha=0.7, position=box_dodge) +

    # 2. Mean points (using separate summary data)
    geom_point(
        data=burrito_means,
        mapping=aes(y='mean', group='Salsa'),
        shape='X',
        size=5,
        position=box_dodge,
        color='black'
    ) +

    # 3. Mean lines
    geom_line(
        data=burrito_means,
        mapping=aes(y='mean', group='Salsa', color='Salsa'),
        size=1.2,
        position=box_dodge
    ) +

    # 4. Facet by Potatoes (two panels side-by-side)
    facet_wrap('~Potatoes', ncol=2) +

    # 5. Styling and labels
    labs(
        title = "Satisfaction by Protein, Salsa, and Potatoes",
        x = "Protein",
        y = "Satisfaction",
        fill = "Salsa",
        color = "Salsa"
    ) +

    theme_classic() +
    theme(
        plot_title = element_text(hjust=0.5),
        strip_text = element_text(size=11, weight='bold')
    ) +

    scale_y_continuous(limits=(0, 15))
)

# Display
p
```

7. Fit the full factorial ANOVA model. Check the assumption of normality using a testing method and a visual method. Does the assumption hold? Why or why not?
* On the QQ plot we can see that the majority of the data point lie on the 45 degree line bisecting the graph with minor devations of approximately 10 percent at the top and bottom ends of the line. This supports the assumption of normality. Additionaly, from the Shapiro-Wilk test we calculate a p-value=0.647 which fails to reject the null hypothesis at an alpha of 0.05. It follows that the statistical test also supports the assumption of normality.

```python
## Import Statsmodels ##
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy import stats

## Fit Full Factorial Model ##

model = ols('Satisfaction ~ Protein*Salsa*Potatoes', data=burrito).fit()

## Parse out model residuals ##

residuals = model.resid

## Test of Normality ##

## QQ-Plot ##

sm.qqplot(residuals, line='s')
plt.title("QQ-Plot of Model Residuals")
plt.show()

## Shapiro-Wilk Test ##

shapiro_test = stats.shapiro(residuals)

print("Shapiro-Wilk Test Results:")
print(f"Statistic: {shapiro_test.statistic}")
print(f"p-value: {shapiro_test.pvalue}")
if shapiro_test.pvalue < 0.05:
    print("Reject the null hypothesis - Residuals are not normally distributed.")
else:
    print("Fail to reject the null hypothesis - Residuals are normally distributed.")
```

8. Check the assumption of constant variance using a testing method and a visual method. Does the assumption hold? Why or why not?
* In the fitted vs. residual plot we can see a random scattering of data points with no clear patterns of coning or clustering. The data points are generally centered around a mean of zero. The banding in the data can be explained by the categorical nature of the variables. Additionally, the Breuch-Pagan test fails to reject the null hypothesis with a p-value=0.678, this further supports the assumption of homoscedasticity.

```python
## Test of Homogeneity of Variance ##

## Scatterplot of Standardized Residuals vs. Fitted Values ##

## Get Standardized Residuals ##

standardized_residuals = model.get_influence().resid_studentized_internal

## Get Fitted Values ##

fitted_values = model.fittedvalues

## Scatterplot ##

## Create the plot ##

plt.figure(figsize=(8, 6))
sns.scatterplot(x=fitted_values, y=standardized_residuals)
plt.axhline(y=0, linestyle='dashed', color='red')
plt.axhline(y=3, color='blue')
plt.axhline(y=-3, color='blue')
plt.xlabel("Predicted y's")
plt.ylabel("Standardized Residuals")
plt.title("Standardized Residuals vs. Fitted Values")
plt.style.use('classic')
plt.show()

## B-P Test ##

## Run B-P Test ##

bp_test = het_breuschpagan(residuals, model.model.exog)

## Extract F-statistic ##

f_statistic = bp_test[2]

## Extract its P-Value ##

f_p_value = bp_test[3]

print(f"F-statistic: {f_statistic}")
print(f"F p-value: {f_p_value}")
if f_p_value < 0.05:
    print("Reject the null hypothesis - Variances are not equal.")
else:
    print("Fail to reject the null hypothesis - Variances are equal.")
```

3. State the independent variables. What lurking variables may be present?
* **Independent Variables**
  * Protein (nominal categorical variable)

  * Potatoes (nominal categorical Variable)

4. Why might a full factorial design be preferred in this specific instance to a $2^k$ fractional factorial design?
* A full factorial design would be prefferable in this instance to a $2^k$ because of the presence of a single category with 3 levels. If we were to choose a $2^k$ design we would lose fidelity in our protein factor for a rather marginal reduction in number of trials required.

5. State all sets of statistical hypotheses.
* For our main effects:
$$ H_{01}: \alpha_{\text{Bacon}} = \alpha_{\text{Sausage}} = \alpha_{\text{Chorizo}} = 0 $$
$$ H_{11}: \text{At least one } \alpha_i \neq 0 $$

$$ H_{02}: \beta_{\text{Mild}} = \beta_{\text{Hot}} = 0 $$
$$ H_{12}: \text{At least one } \beta_j \neq 0 $$

$$ H_{03}: \gamma_{\text{Potatoes}} = \gamma_{\text{No Potatoes}} = 0 $$
$$ H_{13}: \text{At least one } \gamma_k \neq 0 $$



* For our two-way interactions:
$$ H_{04}: (\alpha\beta)_{ij} = 0 \text{ for all } i,j $$
$$ H_{14}: \text{At least one } (\alpha\beta)_{ij} \neq 0 $$

$$ H_{05}: (\alpha\gamma)_{ik} = 0 \text{ for all } i,k $$
$$ H_{15}: \text{At least one } (\alpha\gamma)_{ik} \neq 0 $$

$$ H_{06}: (\beta\gamma)_{jk} = 0 \text{ for all } j,k $$
$$ H_{16}: \text{At least one } (\beta\gamma)_{jk} \neq 0 $$


* For our three-way interaction:
$$ H_{07}: (\alpha\beta\gamma)_{ijk} = 0 \text{ for all } i,j,k $$

6. Perform appropriate exploratory/descriptive analysis including summary statistics and appropriate visualizations. Do these analyses provide any support for the hypotheses? Briefly explain.

    * **Potatoes:** Across the plots, the burritos with potatoes show a small mitigating effect on satisfaction scores. The mean satisfaction of the mild salsa group decreases while the hot salsa increases.

* **Interpreting the Two-Way Interactions:**

    * The slopes of the interaction lines within each potato group are not the same. This suggests a a possible significant two-way interaction effect between Protein and Salso.

* **Interpreting the Three-Way Interaction:**

```python
## Import Relevant Packages ##
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

## Read in burrito Data ##

burrito = pd.read_excel("Busters Burritos.xlsx")

## Data Integrity Check ##

print(burrito.info())

## Exploratory Data Analysis ##
## Start with Main Effects ##

## Protein ##

protien_stats = burrito.groupby("Protein")["Satisfaction"].agg(['mean', 'std'])

print(protien_stats)

## Salsa ##

Salsa_stats = burrito.groupby("Salsa")["Satisfaction"].agg(['mean', 'std'])

print(Salsa_stats)

## Potatoes ##

breath_stats = burrito.groupby("Potatoes")["Satisfaction"].agg(['mean', 'std'])

print(breath_stats)

#two-way interactions
## Protein x Salsa ##

sole_insole_stats = burrito.groupby(["Protein", "Salsa"])["Satisfaction"].agg(['mean', 'std'])

print(sole_insole_stats)

## Protein x Potatoes ##

sole_breath_stats = burrito.groupby(["Protein", "Potatoes"])["Satisfaction"].agg(['mean', 'std'])

print(sole_breath_stats)

## Salsa x Potatoes ##

insole_breath_stats = burrito.groupby(["Salsa", "Potatoes"])["Satisfaction"].agg(['mean', 'std'])

print(insole_breath_stats)

#three-way interactions
sole_insole_breath_stats = burrito.groupby(["Protein", "Salsa", "Potatoes"])["Satisfaction"].agg(['mean', 'std'])

print(sole_insole_breath_stats)

#plotting the data
## Install plotnine
#%pip install plotnine
## Import all plotnine functions ##:
from plotnine import *

# Prepare summarized data
burrito_means = sole_insole_breath_stats.reset_index()[["Protein", "Salsa", "Potatoes", "mean"]]

# Define dodge position for aligning points/lines with the boxplots
box_dodge = position_dodge(0.75)

# Build a single faceted plot
p = (
    ggplot(burrito, aes(x='Protein', y='Satisfaction')) +

    # 1. Boxplot layer (base visualization)
    geom_boxplot(aes(fill='Salsa'), alpha=0.7, position=box_dodge) +

    # 2. Mean points (using separate summary data)
    geom_point(
        data=burrito_means,
        mapping=aes(y='mean', group='Salsa'),
        shape='X',
        size=5,
        position=box_dodge,
        color='black'
    ) +

    # 3. Mean lines
    geom_line(
        data=burrito_means,
        mapping=aes(y='mean', group='Salsa', color='Salsa'),
        size=1.2,
        position=box_dodge
    ) +

    # 4. Facet by Potatoes (two panels side-by-side)
    facet_wrap('~Potatoes', ncol=2) +

    # 5. Styling and labels
    labs(
        title = "Satisfaction by Protein, Salsa, and Potatoes",
        x = "Protein",
        y = "Satisfaction",
        fill = "Salsa",
        color = "Salsa"
    ) +

    theme_classic() +
    theme(
        plot_title = element_text(hjust=0.5),
        strip_text = element_text(size=11, weight='bold')
    ) +

    scale_y_continuous(limits=(0, 15))
)

# Display
p
```
